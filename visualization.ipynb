```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("flats_after_pandas.csv")

max_price = df["Price(USD)"].dropna().max()
max_price_limit = max_price * 0.24

plt.figure(figsize=(10, 8))
plt.hist(df["Price(USD)"].dropna(), bins=30, range=(0, max_price_limit))
plt.xlabel("Ціна (USD)", labelpad=15, fontsize=13)
plt.ylabel("Кількість квартир", labelpad=15, fontsize=13)

plt.title("Розподіл цін на квартири", pad=15, fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df["Total area(m²)"], df["Price(USD)"], color="green", alpha=0.5)
plt.xlabel("Площа (м²)", labelpad=15, fontsize=13)
plt.ylabel("Ціна (USD)", labelpad=15, fontsize=13)
plt.title("Залежність ціни від площі квартири", pad=15, fontsize=15)
plt.show()
```

```python
df_kyiv = df[df["City"] == "Київ"]
df_kyiv = df_kyiv[df_kyiv["Region"].notna() & (df_kyiv["Region"] != "Не відомо")]

avg_price_by_district_kyiv = df_kyiv.groupby("Region")["Price(USD)"].mean().dropna()

plt.figure(figsize=(15, 6))

avg_price_by_district_kyiv.plot(kind="bar", color="coral", fontsize=13)
plt.xlabel("Район", labelpad=15, fontsize=13)
plt.ylabel("Середня ціна (USD)", labelpad=15, fontsize=13)
plt.title("Середня ціна квартир в Києві залежно від району", pad=15, fontsize=20)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
city_counts = df["City"].value_counts()
cities_with_min_50 = city_counts[city_counts >= 50].index
filtered_df = df[df["City"].isin(cities_with_min_50)]
city_avg_price = filtered_df.groupby("City")["Price(USD)"].mean()
top_cities_by_avg_price = city_avg_price.nlargest(10)

plt.figure(figsize=(15, 6))
top_cities_by_avg_price.plot(kind="bar", color="lightblue", fontsize=13)
plt.xlabel("Місто", labelpad=30, fontsize=15)
plt.ylabel("Середня ціна (USD)", labelpad=20, fontsize=15)
plt.title("Топ 10 міст за середньою ціною квартир\n(Кількість оголошень >= 50)", pad=20, fontsize=20)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
city_counts = df["City"].value_counts()
top_cities_by_count = city_counts.head(10)

plt.figure(figsize=(15, 6))
top_cities_by_count.plot(kind="bar", color="lightgreen", fontsize=13)
plt.xlabel("Місто", labelpad=30, fontsize=15)
plt.ylabel("Кількість записів", labelpad=20, fontsize=15)
plt.title("Топ 10 міст за кількістю квартир", pad=20, fontsize=20)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
room_counts = df['Num of rooms'].value_counts()

threshold = 0.05 

small_categories = room_counts[room_counts / room_counts.sum() < threshold]
other_value = small_categories.sum()

room_counts = room_counts[room_counts / room_counts.sum() >= threshold]
room_counts["Інші"] = other_value

room_counts.index = [int(x) if x != "Інші" else x for x in room_counts.index]

plt.figure(figsize=(10, 6)) 
wedges, texts, autotexts = plt.pie(room_counts, autopct="%1.1f%%", startangle=90, colors=plt.cm.Paired.colors, explode=[0.1 if x == "Інші" else 0 for x in room_counts.index])

plt.legend(wedges, room_counts.index, title="Кількість кімнат", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=13)

plt.title("Розподіл квартир за кількістю кімнат", fontsize=15)
plt.tight_layout()
plt.show()
```
