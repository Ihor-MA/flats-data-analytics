```python
import numpy as np
import pandas as pd

df = pd.read_csv('flats.csv')
df.index = np.arange(1, len(df) + 1)
df.head()
```

```python
df.columns = df.columns.str.capitalize()
df.columns = df.columns.str.replace('_', ' ')
df.head()
```

```python
df = df.rename(columns={"Price": "Price(USD)"})
df = df.rename(columns={"Price for square metre": "Price for square metre(USD)"})
df = df.rename(columns={"Total area": "Total area(m²)"})
df.head()
```

```python
df.loc[:, "Price(USD)"] = df["Price(USD)"].str.replace("\u00A0$", "", regex=False)
df.loc[:, "Price(USD)"] = df["Price(USD)"].str.replace(" ", "", regex=False)
df["Price(USD)"] = df["Price(USD)"].astype("int64")
df.head()
```

```python
df["Price for square metre(USD)"] = df["Price for square metre(USD)"].str.replace(" ", "", regex=False)

def different_currency_to_usd(value):
    if "$" in value:
        return int(value.replace("$", ""))
    elif "грн" in value:
        return round(int(value.replace("грн", "").strip()) / 40)
    elif "€" in value:
        return round(int(value.replace("€", "").strip()) * 1.1)
    else:
        return int(value)

df["Price for square metre(USD)"] = df["Price for square metre(USD)"].apply(different_currency_to_usd)
df.head()
```

```python
df["Total area(m²)"] = df["Total area(m²)"].str.replace(" м²", "", regex=False)
df["Total area(m²)"] = df["Total area(m²)"].astype("float64")
df.head()
```

```python
df = df.dropna(subset=["Floor", "Num of floors"])
```

```python
df.loc[:, "Floor"] = df["Floor"].str.replace("Цокольний", "0", regex=False)
df["Floor"] = df["Floor"].astype("int8")
df.dtypes
```

```python
df["Num of floors"] = df["Num of floors"].astype("int8")
df.dtypes
```

```python
df["Num of rooms"] = df["Num of rooms"].astype("Int8")
df.dtypes
```

```python
df.loc[:, "Apartment complex"] = df["Apartment complex"].fillna("Не ЖК")
df.head()
```

```python
def subway_exists_or_not_indicated(row):
    if pd.isna(row["Subway"]):
        if row["City"] in ["Київ", "Харків", "Дніпро"]:
            return "Не вказано"
        else:
            return "Немає в цьому місті"
    else:
        return row["Subway"]


df.loc[:, "Subway"] = df.apply(subway_exists_or_not_indicated, axis=1)
df.iloc[100:110]
```

```python
def fill_rooms(row, df, area_range=10):
    if pd.notna(row["Num of rooms"]):
        return row["Num of rooms"]

    area_min = row["Total area(m²)"] - area_range
    area_max = row["Total area(m²)"] + area_range

    similar_rows = df[(df["Total area(m²)"] >= area_min) & (df["Total area(m²)"] <= area_max)]
    similar_rooms = similar_rows["Num of rooms"].dropna()

    if not similar_rooms.empty:
        return similar_rooms.mean()
    else:
        return np.nan


df = df.copy()

df["Num of rooms"] = df.apply(lambda row: fill_rooms(row, df), axis=1)
df["Num of rooms"] = df["Num of rooms"].round().astype("Int8")
df.iloc[420:430]
```

```python
df.loc[:, "Region"] = df["Region"].fillna("Не відомо")
df.iloc[420:430]
```
